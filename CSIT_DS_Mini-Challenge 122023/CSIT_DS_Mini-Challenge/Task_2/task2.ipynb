```python
%matplotlib inline
import matplotlib.pyplot as plt
import librosa
import librosa.display
```

```python
x, sr = librosa.load('T2_audio_a.wav')
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x[::-1], sr=sr)
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
```

```python
def plot_specshow(filename):
    x, sr = librosa.load(filename)
    X = librosa.stft(x)
    Xdb = librosa.amplitude_to_db(abs(X))
    plt.figure(figsize=(14, 5))
    librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
```

```python
plot_specshow('T2_audio_b.wav')
```

```python
plot_specshow('T2_audio_c.wav')
```

```python
plot_specshow('T2_audio_d.wav')
```
