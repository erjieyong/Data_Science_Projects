# Step 1: Explore physiological signals

This Notebook goes over each of the types of data files containing the original physiological signals for one simulation run, for one subject. The two functions to help you to load and then plot the time series data are:

* outDataFrame = plotData.loadTimeSeries(dataDir, subjectLabel, sessionLabel, runLabel, experimentType, modality)
    * **dataDir**: the top level data directory, i.e. contains the task-ils and task-rest directories
    * **subjectLabel**: the subject descriptor, e.g. sub-cp022
    * **sessionLabel**: the session descriptor, e.g. ses-20210621
    * **runLabel**: the run descriptor, e.g. run-001
    * **experimentType**: the task type, e.g. "task-ils" or "task-rest"
    * **modality**: the keyword in the data file name specifying the physiological modality of the signal(s) it contains, e.g. "lslshimmerrespecg" refers to the ECG and respiration data.
    * returns:
        * **outDataFrame**: a pandas DataFrame object containing the data


* plotData.plotTimeSeries(dataFrame, subjectLabel, sessionLabel, runLabel, dataColumn, dataSampleIdx_Start, dataSampleIdx_End)
    *  **dataFrame**: the pandas DataFrame object containing the data, e.g. as returned by plotData.loadTimeSeries()
    * **subjectLabel, sessionLabel, runLabel**: see above
    * **dataSampleIdx_Start, dataSampleIdx_End**: the start and end indexes of the data to plot.
    
Please refer to the Data Dictionary for detailed information about the physiological modalities in this dataset, as well as their respective sampling rates and data file formats.

## Plot Multiple Streams with Time Synchronization

The first column in the data structures is the common time stamp. The time stamps are stored in the form of Matlab datenum values. We can plot all the signals for a given subject on the common time axis as shown below.

```python
# datenum value of the Unix epoch start (19700101)
t0 = 719529 
day_to_sec = 3600 * 24

# sample conversion to wall-clock datetime
print( "Datetime: " + str(pd.to_datetime(sub022_run001_dfXP11ac['time_dn'][0] - t0, unit = 'D')) )

# collect several time stamp signals and convert to epoch time
t_ac = pd.to_datetime(sub022_run001_dfXP11ac['time_dn'] - t0, unit='D') # Aircraft behavior
t_ed = pd.to_datetime(sub022_run001_dfEDA['time_dn'] - t0, unit='D') # EDA signal
t_pf = pd.to_datetime(sub022_run001_dfXP11pf['time_dn'] - t0, unit='D') # Flight performance
```

```python
# plot elevation
plt.figure(figsize=(10, 6), dpi=100)
plt.subplot(3, 1, 1)
plt.plot(t_ac, sub022_run001_dfXP11ac['aircraft_elevation_m'], label = 'aircraft_elevation_m');
plt.xlabel('Time');
plt.ylabel('Elevation (m)')
plt.xlim([min(t_ac), max(t_ac)])

# plot electrodermal activity
plt.subplot(3, 1, 2)
plt.plot(t_ed, sub022_run001_dfEDA['eda_hand_l_kOhms'], label = 'eda_hand_l_kOhms');
plt.xlabel('Time');
plt.ylabel('Left Hand EDA (kOhms)')
plt.xlim([min(t_ac), max(t_ac)])

# plot total_error 
plt.subplot(3, 1, 3)
plt.plot(t_pf, sub022_run001_dfXP11pf['total_error'], label = 'total_error');
plt.xlabel('Time');
plt.ylabel('Total Error')
plt.xlim([min(t_ac), max(t_ac)])
plt.show()
```
